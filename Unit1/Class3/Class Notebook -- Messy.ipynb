```python
import requests
```

```python
url = 'https://api.census.gov/data.json'
```

```python
req = requests.get(url)
```

```python
req
```

```python
req.text
```

```python
web_data = req.json()
```

```python
web_data['dataset'][0]['c_dataset'][0]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(web_data['dataset'])
```

```python
programCodes = []

for item in web_data['dataset']:
    programCodes.append(item['programCode'][0])
```

```python
programCodes
```

```python
df['programCodes'] = programCodes
```

```python
df.head()
```

```python
item_1 = []
item_2 = []
item_3 = []

for item in web_data['dataset']:
    try:
        item_1.append(item['programCode'][0])
    except:
        item_1.append(None)
        
    try:
        item_2.append(item['programCode'][1])
    except:
        item_2.append(None)
    
    try:
        item_3.append(item['programCode'][2])
    except:
        item_3.append(None)
```

```python
df['programCode1'] = item_1
df['programCode2'] = item_2
df['programCode3'] = item_3
```
