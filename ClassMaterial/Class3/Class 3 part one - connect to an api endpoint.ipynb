```python
import requests
url = 'https://api.census.gov/data.json'
```

```python
req = requests.get(url)
```

```python
req
```

```python
webdata = req.json()
```

```python
webdata.keys()
```

```python
webdata['dataset'][0]
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(webdata['dataset'])
```

```python
programCodes = []

for item in webdata['dataset']:
    programCodes.append(item['programCode'][0])
```

```python
programCodes
```

```python
df['ProgramCodes'] = programCodes
```

```python
df.head()
```

```python
publisher_type = []
publisher_name = []
publisher_subOrganizationOf = []

for item in webdata['dataset']:
    try:
        publisher_type.append(item['publisher']['@type'])
    except:
        publisher_type.append(None)
    
    try:
        publisher_name.append(item['publisher']['name'])
    except:
        publisher_name.append(None)
        
    try:
        publisher_subOrganizationOf.append(item['publisher']['subOrganizationOf'])
    except:
        publisher_subOrganizationOf.append(None)
```

```python
df['PublisherType'] = publisher_type
df['PublisherName'] = publisher_name
df['PublisherSubOrganizationOf'] = publisher_subOrganizationOf
```
