```
#shuffling the data
df=df.sample(frac=1)
```

```
df.head()
```

```
X,y = df['review'],df['sentiment']
y.replace(['positive','negative'],[1,0], inplace=True)
X_train, X_test = X[:40000], X[40000:]
y_train, y_test = y[:40000], y[40000:]
```

```
from tensorflow.keras.preprocessing.text import Tokenizer
```

```
#initialize the tokenizer and set the number of words as 10k
tokenizer = Tokenizer(num_words=10000)
```

```
tokenizer.fit_on_texts(X_train)
```

```
tokenizer.word_index
```

```
X_train = tokenizer.texts_to_sequences(X_train)
X_test = tokenizer.texts_to_sequences(X_test)
```

```
X_train[1]
```

```
from tensorflow.keras.preprocessing.sequence import pad_sequences

X_train = pad_sequences(X_train, maxlen =250)
X_test = pad_sequences(X_test, maxlen=250)
```

```
X_train[0]
```

```
from tensorflow.keras import models
from tensorflow.keras import layers
```

```
mod = models.Sequential([#we store all 10000 words in our list and set the number of weights associated with each word as 64
                         layers.Embedding(10000,64,input_length = 250),
                         layers.Flatten(),
                         layers.Dense(32, activation='relu'),
                         layers.Dense(32, activation='relu'),
                         layers.Dense(1, activation='sigmoid')
                         ])
```

```
mod.summary()
```

```
mod.get_weights()
```

```
mod.compile(optimizer='adam',loss='binary_crossentropy',metrics=['acc'])
```

```
mod.fit(X_train,y_train.values,validation_split=0.2, epochs=5)
```

```
mod.predict(X_test)
```

```
mod.evaluate(X_test, y_test.values)
```
